#### 3.1 - abertura dos dados

```python
#carrega o pandas e le os dados do csv para um pandas.dataframe
import pandas as pd

df = pd.read_csv('./data/breast_cancer.csv')  # --> https://edisciplinas.usp.br/mod/resource/view.php?id=2173861
df.head()
```

#### 3.2 - pré-processamento para limpeza dos dados (se precisar)(conversão de características, conversão de rótulos/classes, imputação de valores faltantes, etc)

```python
#descreve algumas estatisticas do dataframe, utilizadas em conjunto com a df.head() para filtrar colunas
df.describe()
```

```python
#removida coluna sample_id
del df['sample_id']
df.head()
```

```python
#identificando os valores não numericos da coluna diagnosis
df['diagnosis'].unique()
```

```python
#criando o dicionario e fazendo o map para transformar os valores acima em inteiros
d = {
    'malignant' : 0,
    'benign' : 1
    }

df['diagnosis'] = df['diagnosis'].map(d)
df.head()
```

```python
#carregando a coluna diagnosis para o eixo Y
y = df['diagnosis'].values
y
```

```python
#carregando as demais colunas, exceto a sample_id (removida) e diagnosis(30)
X = df.iloc[:, 0:29].values
X
```

#### 3.3 - divisão em treinamento e teste

```python
import numpy as np                                      #utilizando numpy por sugestão da aula
import random                                           #carregando a lib random para ser seed na hora de filtrar amostras
from sklearn.model_selection import train_test_split    #divide os arrays em subconjuntos de treinamento (x_train) e teste (x_test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=random.randint(0,1000), stratify=y)
np.bincount(y_train)                                    #contando elementos separados para treinamento
```

#### 3.4 - normalização dos dados

```python
#utilizando o objeto StandardScaler do pacote sklearn para normalizar os dados
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()   #o valor médio será próximo de 0 e o desvio padrao 1
scaler.fit(X_train)
X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)
X_train_std
```

#### 3.5 - classificação com knn

```python
#implementacao do sklearn dos k-vizinhos mais proximos, com a escolha de k=3
from sklearn.neighbors import KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train_std, y_train)
```

```python
y_pred = knn_model.predict(X_test_std)
print(y_pred)
print(y_test)
```

#### 3.6 - avaliação de acurácia



```python
#avaliação da acuracia utilizando o accuracy_score.
#note que o valor se altera de acordo com a celula [75], visto que foi utilizado a escolha com seed randomico
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
